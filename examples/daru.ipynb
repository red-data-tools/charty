```ruby
require 'charty'

charty = Charty::Plotter.new(:pyplot)
Charty::PyPlot.activate_iruby_integration
```

```ruby
require 'daru'
df = Daru::DataFrame.new({'a':[1,2,3,4], 'b':[4,5,6,7], 'c':[8, 9, 10, 11]})
charty.table = df
```

```ruby
bar = charty.to_bar(:a, :b)
bar.render('sample')
```

```ruby
box_plot = charty.to_box_plot(:a, :b)
box_plot.render('sample')
```

```ruby
bubble = charty.to_bubble(:a, :b, :id)
bubble.render('sample')
```

```ruby
curve = charty.to_curve(:a, :b)
curve.render('sample')
```

```ruby
scatter = charty.to_scatter(:a, :b)
scatter.render('sample')
```

```ruby
error_bar = charty.to_error_bar(:a, :b)
error_bar.render('sample')
```

```ruby
hst= charty.to_hst(:a, :b)
hst.render('sample')
```
